```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#red   = carbonaceous
#black = silicate

Zcent_5AA_sil = 0.65951336
sigma_5AA_sil = 0.6198

Zcent_100AA_sil = 25.25660814
sigma_100AA_sil = 2.2399

Zcent_1000AA_sil = 214.32006646
sigma_1000AA_sil = 6.7213


Zcent_5AA_carb = 1.52701641
sigma_5AA_carb = 0.7857

Zcent_100AA_carb = 44.82365357
sigma_100AA_carb = 3.2455

Zcent_1000AA_carb = 384.12610523
sigma_1000AA_carb = 9.6097
```

```python
################################################################################################
#                                           Silicates
################################################################################################
mu, sigma = Zcent_5AA_sil, sigma_5AA_sil

zmin, zmax  = int(mu - 10*sigma), int(mu + 10*sigma)
ZZ_5AA_sil  = np.arange(zmin, zmax)
ffz_5AA_sil = 1.0/(np.sqrt(2*np.pi*sigma**2))*np.exp(-(ZZ_5AA_sil-mu)**2/(2*sigma**2))


mu, sigma = Zcent_100AA_sil, sigma_100AA_sil

zmin, zmax    = int(mu - 5*sigma), int(mu + 5*sigma)
ZZ_100AA_sil  = np.arange(zmin, zmax)
ffz_100AA_sil = 1.0/(np.sqrt(2*np.pi*sigma**2))*np.exp(-(ZZ_100AA_sil-mu)**2/(2*sigma**2))


mu, sigma = Zcent_1000AA_sil, sigma_1000AA_sil

zmin, zmax     = int(mu - 5*sigma), int(mu + 5*sigma)
ZZ_1000AA_sil  = np.arange(zmin, zmax)
ffz_1000AA_sil = 1.0/(np.sqrt(2*np.pi*sigma**2))*np.exp(-(ZZ_1000AA_sil-mu)**2/(2*sigma**2))

################################################################################################
#                                            Carbonaceous
################################################################################################

mu, sigma = Zcent_5AA_carb, sigma_5AA_carb

zmin, zmax  = int(mu - 10*sigma), int(mu + 10*sigma)
ZZ_5AA_carb  = np.arange(zmin, zmax)
ffz_5AA_carb = 1.0/(np.sqrt(2*np.pi*sigma**2))*np.exp(-(ZZ_5AA_carb-mu)**2/(2*sigma**2))


mu, sigma = Zcent_100AA_carb, sigma_100AA_carb

zmin, zmax    = int(mu - 5*sigma), int(mu + 5*sigma)
ZZ_100AA_carb  = np.arange(zmin, zmax)
ffz_100AA_carb = 1.0/(np.sqrt(2*np.pi*sigma**2))*np.exp(-(ZZ_100AA_carb-mu)**2/(2*sigma**2))


mu, sigma = Zcent_1000AA_carb, sigma_1000AA_carb

zmin, zmax     = int(mu - 5*sigma), int(mu + 5*sigma)
ZZ_1000AA_carb = np.arange(zmin, zmax)
ffz_1000AA_carb = 1.0/(np.sqrt(2*np.pi*sigma**2))*np.exp(-(ZZ_1000AA_carb-mu)**2/(2*sigma**2))
```

```python
xsize = 12
ysize = 5

nfigs_x = 3
nfigs_y = 1

fig = plt.figure(figsize=(xsize, ysize))

hpad = 0.12
wpad = 0.07

xs_panel = 0.90 / nfigs_x 
ys_panel = 0.70 / nfigs_y 

############################################################################################################
############################               5 Angstroms             #######################################
############################################################################################################

ax = fig.add_axes([0, 0, 1, 1])

#ax.plot([0,1], [0,1], visible=False)
ax.set_xlim(0,1)
ax.set_ylim(0,1)

#ax.text(-1.5, 0.9, "5 $\\AA$",   fontsize=20, horizontalalignment='left')

plt.axis('off')

#------------------------------------------------------------------------------------------------
# From bottom left, to upper right.
ii = 0
jj = 0


ax = fig.add_axes([wpad + ii*xs_panel, hpad + jj*ys_panel, xs_panel, ys_panel])


#carbon
ax.plot(ZZ_5AA_carb+0.5, ffz_5AA_carb, "--r", linewidth=2, drawstyle='steps', alpha=1.0)
ax.plot(ZZ_5AA_sil+0.5, ffz_5AA_sil, "-k", linewidth=2, drawstyle='steps', alpha=1.0)


ax.text(2.8, 0.9, "$\\langle$Z$\\rangle =%.1f$"%Zcent_5AA_carb, fontsize=14, color='red')
ax.text(2.8, 0.8, "$\\sigma_{Z}=%.1f$"%sigma_5AA_carb, fontsize=14, color='red')

ax.text(0.8, 0.9, "$\\langle$Z$\\rangle =%.1f$"%Zcent_5AA_sil, fontsize=14, color='black')
ax.text(0.8, 0.8, "$\\sigma_{Z}=%.1f$"%sigma_5AA_sil, fontsize=14, color='black')

ax.set_ylim(0, 0.99)
ax.set_xlim(-1.78,4.8)

ax.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='on',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='on') # labels along the bottom edge are off

ax.set_ylabel("f(Z)", fontsize=20)
ax.set_xlabel("Z", fontsize=18)

ax.tick_params(axis='both', which='major', length=4, width=2, labelsize=15)
ax.text(-1.5, 0.9, "5 $\\AA$",   fontsize=20, horizontalalignment='left')


############################################################################################################
############################                100 Angstroms            #######################################
############################################################################################################

#------------------------------------------------------------------------------------------------
# From bottom left, to upper right.
ii = 0
jj = 1


ax = fig.add_axes([wpad + jj*xs_panel, hpad + ii*ys_panel, xs_panel, ys_panel])

ax.plot(ZZ_100AA_carb+0.5, ffz_100AA_carb*2, "--r", linewidth=2, drawstyle='steps', alpha=1.0)
ax.plot(ZZ_100AA_sil+0.5, ffz_100AA_sil*2, "-k", linewidth=2, drawstyle='steps', alpha=1.0)


ax.text(43, 0.9, "$\\langle$Z$\\rangle =%.1f$"%Zcent_100AA_carb, fontsize=14, color='red')
ax.text(43, 0.8, "$\\sigma_{Z}=%.1f$"%sigma_100AA_carb, fontsize=14, color='red')
ax.text(43, 0.7, "$f\,(Z) \\times 2$", fontsize=14, color='red')

ax.text(26, 0.9, "$\\langle$Z$\\rangle =%.1f$"%Zcent_100AA_sil, fontsize=14, color='black')
ax.text(26, 0.8, "$\\sigma_{Z}=%.1f$"%sigma_100AA_sil, fontsize=14, color='black')
ax.text(26, 0.7, "$f\,(Z) \\times 2$", fontsize=14, color='black')


#ax.set_xlim(np.min(Z_a_3_100)+0.1, np.max(Z_a_3_100)-0.1)
ax.set_ylim(0, 0.99)
ax.set_xlim(5, 61)

ax.tick_params(
    axis='both',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='on',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='on', labelleft='off') # labels along the bottom edge are off


ax.set_xlabel("Z", fontsize=18)

ax.tick_params(axis='both', which='major', length=4, width=2, labelsize=15)
ax.text(14., 0.9, "100 $\\AA$", fontsize=20, horizontalalignment='center')




############################################################################################################
############################               1000 Angstroms            #######################################
############################################################################################################

#------------------------------------------------------------------------------------------------
# From bottom left, to upper right.
ii = 0
jj = 2



ax = fig.add_axes([wpad + jj*xs_panel, hpad + ii*ys_panel, xs_panel, ys_panel])


ax.plot(ZZ_1000AA_carb+0.5, ffz_1000AA_carb*5, "--r", linewidth=2, drawstyle='steps', alpha=1.0)
ax.plot(ZZ_1000AA_sil+0.5, ffz_1000AA_sil*5, "-k", linewidth=2, drawstyle='steps', alpha=1.0)

ax.text(330, 0.9, "$\\langle$Z$\\rangle =%.1f$"%Zcent_1000AA_carb, fontsize=14, color='red')
ax.text(330, 0.8, "$\\sigma_{Z}=%.1f$"%sigma_1000AA_carb, fontsize=14, color='red')
ax.text(330, 0.7, "$f\,(Z) \\times 5$", fontsize=14, color='red')

ax.text(245, 0.9, "$\\langle$Z$\\rangle =%.1f$"%Zcent_1000AA_sil, fontsize=14, color='black')
ax.text(245, 0.8, "$\\sigma_{Z}=%.1f$"%sigma_1000AA_sil, fontsize=14, color='black')
ax.text(245, 0.7, "$f\,(Z) \\times 5$", fontsize=14, color='black')


ax.set_ylim(0, 0.99)
ax.set_xlim(155, 410)

ax.tick_params(
    axis='both',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='on',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='on', labelleft='off') # labels along the bottom edge are off


ax.set_xlabel("Z", fontsize=18)

ax.tick_params(axis='both', which='major', length=4, width=2, labelsize=12)
ax.text(200., 0.9, "1000 $\\AA$", fontsize=20, horizontalalignment='center')


fig.show()

fig.savefig("/home/jcibanezm/Dropbox/Projects/Papers/In_Progress/ChargeDist_Dust/Figures/PDR_Gaussian_fZ.pdf", format="pdf")
```
